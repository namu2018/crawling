### 만약에 태그에 속성이 있다면 속성값은 어떻게 가져올까?


```
from bs4 import BeautifulSoup

html = """
<html>
<body>
<a href="https://www.naver.com" style="color:blue">naver</a>
<p>test</p>
</body>
</html>"""

soup = BeautifulSoup(html, 'lxml')
```

```
soup.a.text
```

```
inf = soup.a
print(inf.attrs)
```

```
inf['href']
```

```
inf['style']
```

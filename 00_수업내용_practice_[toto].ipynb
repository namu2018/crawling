```
### 만약에 Timeout 이 걸렸다.
import requests as rq
import time

url = "https://github.com/LDJWJ/DeepLearning_Class"
delay_time = 1

def connection(u):
  return rq.get(u)

try:
  connection(url)  
except rq.exceptions.Timeout:
  time.sleep(delay_time)
  connection(url)
except rq.exceptions.ReadTimeout:
  print("서버가 일정시간동안 데이터를 보내지 않았습니다.")
except rq.exceptions.MissingSchema:
  print("HTTP 또는 HTTPS를 생략했습니다.")
except rq.exceptions.InvalidURL:
  print("URL이 잘못됐습니다.")
except rq.exceptions.ConnectTimeout:
  print("일정시간동안 서버가 응답하지 않았습니다")
except rq.exceptions.TooManyRedirects:
  print("리다이렉션이 너무 많습니다.")
  
```

```
### urllib 사용 - 기타 모듈
from urllib.request import urlopen, Request

url = "https://github.com/LDJWJ/DeepLearning_Class"
req = Request(url)
page = urlopen(req)

print(page)
print(page.code)
print(page.headers)
print(page.url)
print(page.info().get_content_charset())
```
